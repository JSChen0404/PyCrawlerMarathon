### ① 取出今天所有的發文

```python
import requests, time
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'

#request的连接数过多而导致Max retries exceeded
#在header中不使用持久连接

headers = {'Connection': 'close'}
r = requests.get(url, headers=headers)

soup = BeautifulSoup(r.text, "html.parser")
soup
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# soup.find(class_="part_menu_2").find_all('a').text
list_1 =[]
dirs = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    cate = d.find('em').text
    title = d.find('a').text
    list_1.append(title)
    if cate not in dirs.keys():
        dirs[cate] = [title]
    else:
        dirs[cate].append(title)

print(len(list_1),'\n')
print(dirs['政治'])
```

### ③ 哪一個類別的文章最多

```python
for key, value in dirs.items():
    print(key,' : ',len(value))
# 影劇
```
