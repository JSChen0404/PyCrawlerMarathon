### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
# 沒加 header
url = r'https://www.zhihu.com/explore'
r = requests.get(url, verify=False)
response = r.text
response
```

```python
# 加 header
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get(url, 'utf-8', headers=headers)
response = r.text
response
```
