### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
browser = webdriver.Chrome(executable_path=r'C:\Users\user\Dropbox\23 - Python\Program\chromedriver.exe')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
soup
```

```python
# 1. 先取出全部有幾頁
for tag in soup.find_all('div'):
    if tag.get('homepagerandwindow'):
        print("總共有",tag.get('homepagerandwindow'), "頁")
```

```python
browser = webdriver.Chrome(executable_path=r'C:\Users\user\Dropbox\23 - Python\Program\chromedriver.exe')
browser.get("https://hahow.in/courses?page="+str(1))
```

```python
# 2. 每頁都取出資料放入 courses
browser = webdriver.Chrome(executable_path=r'C:\Users\user\Dropbox\23 - Python\Program\chromedriver.exe')

courses = []
for i in range(1,33):
    browser.get("https://hahow.in/courses?page="+str(i))
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    for d in soup.find_all('h4'):
        courses.append(d.text)   
browser.close()
```

```python
# 3. done!
print(courses)
```
