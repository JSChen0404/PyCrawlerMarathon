## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
import requests
from bs4 import BeautifulSoup
import random
import time
```

```python
r = requests.get('https://free-proxy-list.net')
soup = BeautifulSoup(r.text, "html.parser")
soup
```

```python
proxy_ips = []
rows = soup.find("table", class_="table table-striped table-bordered" ).tbody.find_all('tr')
for row in rows:
    ip = row.find_all('td')
    ip_combine = ip[0].text + ':' + ip[1].text
#     print(ip_combine)
    proxy_ips.append(ip_combine)

print(proxy_ips[0:5])
#此網站總共有300個ip
```

```python
for i in range(10):
    ip = random.choice(proxy_ips)
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php', verify=False,
                        proxies={'http': ip, 'https': ip}, timeout=10)
        soup = BeautifulSoup(resp.text, 'html.parser')
        print(soup.find('h1', id='ipd').text.strip())
    except Exception as f:
        print('Failed result:\n',f)
        pass
    
    n = random.choice(range(4,10))
    if i==9:
        print('-'*100)
        print('Done')
    else:
        print('wait for {} seconds...'.format(n))
        print()
        time.sleep(n)
```
